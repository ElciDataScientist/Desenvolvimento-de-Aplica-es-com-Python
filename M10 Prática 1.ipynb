```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python

N = 50
x = 0 + 8*np.random.rand(N)
y = -1*x + np.random.randn(N)*.4
df8 = pd.DataFrame({'x':x, 'y':y})

_ = sns.regplot(x='x', y='y', data = df8)
df8.corr()
```

```python
df8.iloc[10,0] = 10
df8.iloc[10,1] = 15

_ = sns.regplot(x='x', y='y', data = df8)
plt.xlim(-.5, 10.5)
plt.ylim(-10, 17)
df8.corr()
```

```python
x[10] = np.nan
y[10] = np.nan
_ = sns.regplot(x=x, y=y)
np.corrcoef(x,y)[1,0]
```
