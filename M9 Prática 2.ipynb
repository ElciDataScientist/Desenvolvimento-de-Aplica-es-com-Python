```python
data = {
    'munResLat': [-11.0, -10.0, -10.7, -9.5, -11.3, -10.6, -12.0, -9.0],
    'variavel1': [100, 200, 150, 300, 120, 180, 220, 290],
    'variavel2': [15, 25, 20, 35, 18, 28, 22, 30]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['faixa_latitude'] = pd.cut(df['munResLat'], bins=[-np.inf, -10.5, np.inf], labels=['Abaixo -10.5', 'Acima -10.5'])
```

```python
resultados = df.groupby('faixa_latitude').agg({
    'variavel1': ['sum', 'mean', 'min', 'max', 'median', 'std', 'var'],
    'variavel2': ['sum', 'mean', 'min', 'max', 'median', 'std', 'var']
})
```

```python
df_mulheres['data'] = pd.to_datetime(df_mulheres['data'])
```

```python
df_mulheres['ano_mes'] = df_mulheres['data'].dt.to_period('M')
idade_media_por_regiao = df_mulheres.groupby(['ano_mes', 'regiao_imediata'])['idade'].mean().unstack()
```

```python
plt.figure(figsize=(14, 8))
for regiao in idade_media_por_regiao.columns:
    plt.plot(idade_media_por_regiao.index.astype(str), idade_media_por_regiao[regiao], label=regiao)
```

```python
# Converta 'ano_mes' em string antes de plotar:
df_mulheres['ano_mes'] = df_mulheres['ano_mes'].astype(str)

# Plotando o gráfico
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_mulheres, x='ano_mes', y='idade', hue='regiao_imediata', marker='o')

# Configurações do gráfico
plt.xlabel('Data')
plt.ylabel('Idade Média das Mulheres')
plt.title('Idade Média das Mulheres ao Longo do Tempo por Região Imediata de Rondônia')
plt.xticks(rotation=45)
plt.legend(title='Região Imediata')
plt.grid(True)
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
